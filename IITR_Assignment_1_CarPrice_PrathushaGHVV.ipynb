```python
# USING LABELENCODER TO CONVERT OBJECT TYPE ENTRY TO INT TYPE

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cp[cp.select_dtypes(include=['object']).columns]  =  cp[cp.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
cp.head()
```

```python
# SPLITTING THE DATASET INTO TEST AND TRAIN

from sklearn.model_selection import train_test_split
```

```python
cp_x = cp.iloc[:, 0:24]

cp_y = cp.iloc[:, 24]

x_train, x_test, y_train, y_test = train_test_split(cp_x, cp_y, train_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
predict = lr.predict(x_train)
```

```python
# SLOPE /m

lr.coef_
```

```python
x_train.columns
```

```python
# Intercept

lr.intercept_
```

```python
#RSQUARE

Rsquar = lr.score(x_train, y_train)

Rsquar
```

```python
# ADJUSTED R-SQUARE

N = x_train.shape[0]

K = x_train.shape[1]

Rsquar = lr.score(x_train, y_train)

ADJRsquar = 1-((1-Rsquar)*(N-1)/(N-K-1))

ADJRsquar
```

```python
train_prediction = lr.predict(x_train)
```

```python
test_prediction = lr.predict(x_test)
```

```python
# TRAIN ERROR

error = y_train - train_prediction

error
```

```python
# HISTOGRAM ON TRAIN ERROR

import matplotlib.pyplot as plt

import seaborn as sns

plt.hist(error);
```

```python
# DISTPLOT OF TRAIN ERROR

sns.distplot(error);
```

```python
# TEST ERROR

error_test = y_test - test_prediction
```

```python
error_test.median()
```

```python
# MSE

import numpy as np

MSE = np.median(np.square(error_test))

MSE
```

```python
# RMSE

RMSE = np.sqrt(error_test)

RMSE.median()
```

```python
# MAPE

MAPE = np.median(np.abs(error_test.median()/y_test.median()))

MAPE
```

```python
# ACCURACY

ACCURACY = 100-MAPE*100

ACCURACY
```
